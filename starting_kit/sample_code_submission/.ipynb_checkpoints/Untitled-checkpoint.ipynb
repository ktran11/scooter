```python
import numpy as np# We recommend to use numpy arrays
from os.path import isfile
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import IsolationForest
import pandas as pd # for using pandas daraframe
from sklearn.preprocessing import StandardScaler # for standardizing the Data
from sklearn.feature_selection import GenericUnivariateSelect, f_regression
from sklearn.ensemble import RandomForestRegressor
from sklearn.decomposition import PCA

class model ():

    def __init__(self):
  
        self.num_train_samples= 38563
        self.num_feat=59
        self.num_labels=1
        self.is_trained=False
        self.preprocess = GenericUnivariateSelect()
        self.mod = RandomForestRegressor(n_estimators=110, max_depth = 30,min_samples_split=5) 
    
    
    def preprocessing(self,data):
        clf = IsolationForest( behaviour = 'new', max_samples='auto', random_state = 1, contamination= 'auto')
        preds = clf.fit_predict(data)
        preds = np.where( preds<1)
        for i in preds: 
            data = data.drop(i,axis=0)
        
        scaler = StandardScaler()
        scaler.fit(data)
        StandardScaler(copy=True ,with_mean=True ,with_std =True )
        scaled_data =scaler.transform(data)

        pca = PCA(0.99)
        pca.fit(scaled_data)
        x_pca =pca. transform(scaled_data)
        
        
        X = data[data.columns[0:len(data.columns)-1]].values 
        Y = data[data.columns[-1]].values
        initial_feature_names = data.columns
        transformer = GenericUnivariateSelect(f_regression, 'k_best', param=pca.n_components_)
        X_new = transformer.fit_transform(X, Y[:,-1])
        return X_new
    
    
    def fit(self, X, y):
        df = pd.DataFrame(X)
        target= pd.DataFrame(y)
        data = pd.concat([df,target],axis=1)
        
        X= self.preprocessing(data)
       
        if X.ndim>1: self.num_feat = X.shape[1]
        if y.ndim>1: self.num_labels = y.shape[1]
        
        #X_preprocess= self.preprocess.fit_transform(X)
        self.mod.fit(X, y)
        self.is_trained = True

    def predict(self, X):
        num_test_samples = X.shape[0]
       
        y = np.zeros(num_test_samples,)
        df = pd.DataFrame(X)
        target= pd.DataFrame(y)
        
        data = pd.concat([df,target],axis=1)
        
        X= self.preprocessing(data)
        if X.ndim>1: num_feat = X.shape[1]

        y = self.mod.predict(X)
        return y

    def save(self, path="./"):
        pass

    def load(self, path="./"):
        pass


def test():
    mod = model()
    X_random = np.random.rand(mod.num_train_samples,mod.num_feat)
    Y_random = np.random.rand(mod.num_train_samples)
    mod.fit(X_random,Y_random)
    mod.predict(X_random)

if __name__ == "__main__":
    test()
```
